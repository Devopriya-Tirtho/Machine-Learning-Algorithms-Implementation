```python
import io
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('blobs.txt',sep=" ",header = None)
#print(data)
a=[it for it in data[0]]
b=[it for it in data[1]]
plt.plot(a ,b ,marker="*",linestyle = 'None',color="blue")
```

```python
k = int(input("Enter your value: "))
import random 
#k=3;
datalen=len(data[0])
print(datalen)
centroids=[]
for i in range(k):
  index=random.randint(0,datalen-1)
  centroids.append([data[0][index],data[1][index]])
print(centroids)
```

```python
import math
lst=[]
converg=False
for m in range (0,200):
  
  if(m>0):
    centroidsTmp=[]
    for j in range(k):
      classwisedata1=[it[0] for it in lst if it[2]==j]
      classwisedata2=[it[1] for it in lst if it[2]==j]
     
      
      centroidsTmp.append([sum(classwisedata1)/len(classwisedata1),sum(classwisedata2)/len(classwisedata2)])
    
    if(centroids==centroidsTmp):
      converg=True
    else:
      centroids=centroidsTmp
      
  if(converg):
    break
  lst=[]
  print("iter ",m)
 # print(centroids)
  for i in range(0,datalen):
    dist=[]
    for j in range(k):
      dist.append(math.sqrt(pow(centroids[j][0]-data[0][i],2)+pow(centroids[j][1]-data[1][i],2)))
      
    #print(dist.index(min(dist)))
    lst.append([data[0][i],data[1][i],dist.index(min(dist))])  
  
  print(lst) 
```
